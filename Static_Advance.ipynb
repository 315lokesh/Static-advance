Que.14 How does the Central Limit Theorem relate to the normal distribution?

Ans: The Central Limit Theorem (CLT) states that the distribution of sample means will approach a normal distribution as the sample size increases, regardless of the shape of the original population distribution. In essence, it links the concept of sampling distributions to the familiar normal distribution.

**Here's a more detailed explanation:**

**Sampling Distribution:**
The CLT focuses on the distribution of sample means, which is called the sampling distribution. Imagine repeatedly drawing samples from a population and calculating the mean of each sample. The distribution of all these sample means is the sampling distribution.

**Approaching Normality:**
The CLT asserts that, regardless of the original population's distribution (which could be skewed, uniform, etc.), the sampling distribution will become increasingly normal as the sample size (number of observations in each sample) gets larger.

**Normal Distribution as a Limit:**
The "limit" in "Central Limit Theorem" refers to the fact that the sampling distribution approaches a normal distribution as the sample size approaches infinity.

**Why it Matters:**
The CLT is crucial because it allows us to use many statistical methods that assume normality, even when the original data is not normally distributed. For example, t-tests and ANOVA can be applied with reasonable accuracy to non-normal data if the sample size is large enough due to the CLT, according to Statsig.

**Example:**
If you were to repeatedly take samples of size 30 from a population of exam scores (which might not be normally distributed), the distribution of the average exam scores for those samples would be approximately normal.
In short, the Central Limit Theorem provides a powerful link between sample means and the normal distribution, making it a cornerstone of statistical inference.

Que.17 What are point estimates and interval estimates in statistics?

Ans: n statistics, point estimates and interval estimates are two methods used to estimate unknown population parameters, but they differ in approach. A point estimate provides a single, specific value as the best guess for a parameter, such as using the sample mean to estimate a population mean. In contrast, an interval estimate gives a range of values that is likely to contain the population parameter, often expressed as a confidence interval.

In this article, we will discuss about point and interval estimate. We will also discuss key differences between Point and Interval Estimate in detail.

What is a Point Estimate?
A point estimate is a single value used to approximate an unknown population parameter. It is derived from sample data and provides the best guess for the true value of a population characteristic, such as the mean or proportion.

For example:

The sample mean is a point estimate of the population mean.
The sample proportion is a point estimate of the population proportion.

What is an Interval Estimate?
An interval estimate is a range of values, derived from sample data, that is used to estimate an unknown population parameter. Unlike a point estimate, which gives a single value, an interval estimate provides a range within which the true value of the parameter is likely to fall, offering a degree of uncertainty.

The most common type of interval estimate is a confidence interval, which specifies a range around the point estimate and includes a confidence level (typically 90%, 95%, or 99%). This confidence level represents the probability that the interval contains the true population parameter.

For example:

If we calculate a 95% confidence interval for the population mean, it means that if we were to take many samples and calculate confidence intervals for each, 95% of those intervals would contain the true population mean.

Que.21 What are the assumptions for applying the Central Limit Theorem?

Ans: What Is the Central Limit Theorem (CLT)?The Central Limit Theorem (CLT) requires a few key assumptions to be met for the sampling distribution of sample means to approximate a normal distribution. These include: random sampling, independence of samples, a sufficiently large sample size, and sometimes, that the sample size is less than 10% of the population when sampling without replacement.

**Here's a more detailed explanation:**

Random Sampling:
The sample must be drawn randomly from the population. This ensures that each member of the population has an equal chance of being selected, which is crucial for reducing bias and making the sample representative of the population.

Independence of Samples:
The individual observations within the sample should be independent of each other. This means that the selection of one data point should not influence the selection or value of any other data point in the sample.

Sufficiently Large Sample Size:
The sample size needs to be large enough for the CLT to be effective. While a rule of thumb is often cited as n ≥ 30, the necessary sample size can vary depending on the shape of the population distribution. For distributions that are heavily skewed or have outliers, a larger sample size might be needed.

Sampling Without Replacement:
If sampling is done without replacement, the sample size should not exceed 10% of the total population. This helps to maintain the independence of the samples.

It's a weighted average of all possible values, where each value is weighted by its corresponding probability. Essentially, it tells you what you can "expect" to happen on average if you repeat an experiment many times.

Here's a more detailed explanation:

Probability Distribution:
A probability distribution describes the likelihood of each possible outcome of a random variable.

Random Variable:
A random variable is a variable whose value is a numerical outcome of a random phenomenon.

Expected Value (E(X)):
For a discrete random variable, the expected value is calculated by multiplying each possible outcome by its probability and summing the results. The formula is: E(X) = Σ [x * P(x)], where 'x' represents each outcome and 'P(x)' its probability.

Interpreting Expected Value:
The expected value is not necessarily a value that will be observed in any single trial of the experiment. Instead, it represents the average outcome over many repetitions. If you were to repeatedly run the experiment, the average of the outcomes would tend towards the expected value

Expected Value:

Relationship:
The probability distribution provides the probabilities needed to calculate the expected value. The expected value, in turn, summarizes the entire distribution into a single, representative value, indicating the average outcome you would expect in the long run.

Example:

The expected value, in this case, would be (1 * 0.6) + (0 * 0.4) = 0.6, meaning you would expect to get heads on average 60% of the time if you flipped the coin many times.

```python
#Que.2  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define range (like a dice: 1 to 6)
low = 1
high = 7  # upper bound is exclusive

# Create the distribution
dist = randint(low, high)

# Outcomes and PMF
x = np.arange(low, high)
pmf_values = dist.pmf(x)

# Plot PMF
plt.figure(figsize=(8, 5))
plt.stem(x, pmf_values)  # use_line_collection removed for compatibility
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.xticks(x)
plt.ylim(0, 0.3)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

```

```python
#Que.3 +> Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): 0 or 1 (possible outcomes)
    p (float): probability of success (0 <= p <= 1)

    Returns:
    float: Probability P(X = x)
    """
    if x not in [0, 1]:
        return "Invalid input: x must be 0 or 1"
    if not (0 <= p <= 1):
        return "Invalid input: p must be between 0 and 1"

    return p**x * (1 - p)**(1 - x)

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Success
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Failure

```

```python
#Que.6  Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution

import numpy as np
import matplotlib.pyplot as plt

# Define the range of the discrete uniform distribution
a = 1  # minimum value
b = 6  # maximum value (like a dice roll)
x = np.arange(a, b + 1)

# Probability Mass Function (PMF)
pmf = np.full_like(x, 1 / (b - a + 1), dtype=np.float64)

# Cumulative Distribution Function (CDF)
cdf = np.cumsum(pmf)

# Plotting CDF
plt.step(x, cdf, where='post', color='blue', label='CDF')
plt.scatter(x, cdf, color='red')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(x)
plt.ylim(0, 1.05)
plt.legend()
plt.show()
```

```python
#Que.9
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Step 1: Calculate mean and std deviation
    mean = np.mean(data)
    std = np.std(data)

    # Step 2: Calculate Z-scores
    z_scores = (data - mean) / std

    # Step 3: Plot histogram of Z-scores
    plt.hist(z_scores, bins=50, color='skyblue', edgecolor='black', density=True)
    plt.title('Z-score Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data = np.random.normal(loc=50, scale=10, size=1000)  # Normally distributed data
z_scores = calculate_and_plot_zscores(data)
```

```python
#Que.14
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    return z

def plot_standard_normal_with_z(z_score):
    # Generate x values for standard normal
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    # Plot standard normal curve
    plt.plot(x, y, color='blue', label='Standard Normal PDF')
    plt.fill_between(x, y, alpha=0.2, color='lightblue')

    # Mark the Z-score on the curve
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
    plt.title('Z-score Comparison with Standard Normal Distribution')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example input
data_point = 75
mean = 60
std_dev = 10

# Step 1: Calculate Z-score
z = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point} is: {z:.2f}")

# Step 2: Plot
plot_standard_normal_with_z(z)
```

```python
#18
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mu, sigma):
    return (x - mu) / sigma

def plot_standard_normal_with_z(z_score):
    # Create x values from -4 to 4
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color='blue')

    # Shade area under the curve up to z_score
    x_shade = np.linspace(-4, z_score, 1000)
    y_shade = norm.pdf(x_shade)
    plt.fill_between(x_shade, y_shade, color='skyblue', alpha=0.5, label=f'Area ≤ Z={z_score:.2f}')

    # Draw a line at the z-score
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')

    plt.title("Z-score in Standard Normal Distribution")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: Given data point
x = 75          # data point
mu = 70         # population mean
sigma = 5       # standard deviation

# Calculate Z-score
z = calculate_z_score(x, mu, sigma)
print(f"Z-score for X = {x} is: {z:.2f}")

# Plot it
plot_standard_normal_with_z(z)

```

```python
#19
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = np.array([72, 74, 75, 70, 71, 69, 68, 74, 73, 75,
                        71, 72, 73, 74, 75, 72, 71, 73, 74, 75,
                        70, 72, 73, 74, 75, 73, 72, 71, 74, 75,
                        73, 74, 75, 70, 72, 74, 75, 72, 73, 74,
                        73, 71, 70, 74, 75, 72, 73, 74, 75, 72])

# Known population parameters
mu_0 = 70      # hypothesized population mean
sigma = 10     # known population standard deviation

# Sample stats
x_bar = np.mean(sample_data)
n = len(sample_data)

# Z-statistic
z = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Significance level
alpha = 0.05

# Results
print(f"Sample mean: {x_bar:.2f}")
print(f"Z-statistic: {z:.3f}")
print(f"P-value: {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
#20
import numpy as np
from scipy.stats import norm

# Sample data (you can use your own dataset)
data = np.array([72, 74, 75, 70, 71, 69, 68, 74, 73, 75,
                 71, 72, 73, 74, 75, 72, 71, 73, 74, 75,
                 70, 72, 73, 74, 75, 73, 72, 71, 74, 75])

# Step 1: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)   # Sample standard deviation
n = len(data)
confidence = 0.95

# Step 2: Calculate Z-critical value
z_critical = norm.ppf(1 - (1 - confidence) / 2)  # For 95% CI

# Step 3: Calculate Margin of Error
margin_of_error = z_critical * (sample_std / np.sqrt(n))

# Step 4: Calculate Confidence Interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Output
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Confidence Interval (95%) = [{lower_bound:.2f}, {upper_bound:.2f}]")
```

```python
#21
import numpy as np
from scipy.stats import norm

# Step 1: Generate random data from a normal distribution
np.random.seed(42)            # for reproducibility
mu = 100                      # population mean
sigma = 15                    # population standard deviation
n = 50                        # sample size

sample = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # sample standard deviation (with Bessel's correction)

# Step 3: Calculate Z-critical value for 95% confidence
confidence_level = 0.95
z_critical = norm.ppf(1 - (1 - confidence_level) / 2)  # ≈ 1.96

# Step 4: Calculate Margin of Error
margin_of_error = z_critical * (sample_std / np.sqrt(n))

# Step 5: Confidence Interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 6: Print Results
print(f"Sample Mean           = {sample_mean:.2f}")
print(f"Sample Std Dev        = {sample_std:.2f}")
print(f"Z-critical value      = {z_critical:.3f}")
print(f"Margin of Error       = {margin_of_error:.2f}")
print(f"95% Confidence Interval = [{lower_bound:.2f}, {upper_bound:.2f}]")
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Define the lambda (average number of events)
lam = 4  # mean (λ) of the distribution

# Step 2: Define x values (discrete values)
x = np.arange(0, 15)

# Step 3: Calculate CDF values
cdf_values = poisson.cdf(x, mu=lam)

# Step 4: Print CDF values
print("x\tP(X ≤ x)")
for xi, cdf in zip(x, cdf_values):
    print(f"{xi}\t{cdf:.4f}")

# Step 5: Plot the CDF
plt.figure(figsize=(10, 5))
plt.step(x, cdf_values, where='post', color='purple', label=f'Poisson CDF (λ={lam})')
plt.title('Poisson Cumulative Distribution Function (CDF)')
plt.xlabel('x (Number of Events)')
plt.ylabel('P(X ≤ x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#28
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mu=0, sigma=1, size=1000):
    # Step 1: Generate data from normal distribution
    data = np.random.normal(loc=mu, scale=sigma, size=size)

    # Step 2: Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # excess kurtosis (normal → 0)

    # Step 3: Print results
    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (Excess): {data_kurtosis:.4f}")

    # Optional: Return for further analysis
    return data_skewness, data_kurtosis

# Example usage
analyze_normal_distribution_skew_kurtosis(mu=0, sigma=1, size=1000)
```
